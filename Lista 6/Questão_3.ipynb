```python
# Import all necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc, classification_report
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV
from scipy.special import expit
from sklearn.preprocessing import label_binarize
from sklearn.naive_bayes import GaussianNB
```

```python
# generate 2 class dataset
x, y = make_classification(n_samples=10000, n_classes=2, weights=[0.9,0.5],
random_state=42)

# Split data into train and test subsets.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8)

```

```python
idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)

fig = plt.figure(figsize=(5,5))
plt.plot(x[idx0,0], x[idx0,1], '.', label='Class 0')
plt.plot(x[idx1,0], x[idx1,1], 'rx', label='Class 1')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.show()
```

```python
# Instantiate LogisticRegression object for multi-class case.
model = LogisticRegression(solver='lbfgs', max_iter=10000, multi_class='multinomial')

# Train model.
model.fit(x_train, y_train)

# Predict.
y_pred = model.predict(x_test)
```

```python
# Instantiate a Gaussian naive Bayes classifier.
gnb = GaussianNB()

# fit.
gnb.fit(x, y)

#prediction
y_pred2 = gnb.predict(x_test)
```

```python
#GridSearch

parameters = { 'weights' :( 'uniform' , 'distance' ), 'n_neighbors' :[1,2,3,4,5,10,15,20]}
svc = KNeighborsClassifier()
clf = GridSearchCV(svc, parameters, cv=10, verbose=3, n_jobs=-1)
clf . fit(x , y)
clf .best_params_
clf .best_score_
print(clf .best_params_)
print(clf. best_score_)
```

```python
# Create an instance of Neighbours Classifier and fit the data with best k.
knc = KNeighborsClassifier(n_neighbors=15, weights='uniform')
# Train the classifier.
knc.fit(x_train, y_train)
# Predict.
y_pred_knn = knc.predict(x_test)
```

```python
# Getting the probabilities for each class.
y_prob_log = model.predict_proba(x_test)
y_prob_gnb = gnb.predict_proba(x_test)
y_prob_knn = knc.predict_proba(x_test)

# Binarize the test targets.
y_test_bin_log = label_binarize(y_test, classes=[0, 1, 2])
y_test_bin_gnb = label_binarize(y_test, classes=[0, 1, 2])
y_test_bin_knn = label_binarize(y_test, classes=[0, 1, 2])

# Compute ROC curve and ROC area for each class.
fpr_log = dict()
tpr_log = dict()
roc_auc_log = dict()

fpr_gnb = dict()
tpr_gnb = dict()
roc_auc_gnb = dict()

fpr_knn = dict()
tpr_knn = dict()
roc_auc_knn = dict()

plt.figure(figsize=(14, 5))

lw = 2

for i in range(2):
    # Calculating ROC curve and ROC AUC.
    fpr_log[i], tpr_log[i], _ = roc_curve(y_test_bin_log[:, i], y_prob_log[:, i])
    roc_auc_log[i] = auc(fpr_log[i], tpr_log[i])
    
    fpr_gnb[i], tpr_gnb[i], _ = roc_curve(y_test_bin_gnb[:, i], y_prob_gnb[:, i])
    roc_auc_gnb[i] = auc(fpr_gnb[i], tpr_gnb[i])
    
    fpr_knn[i], tpr_knn[i], _ = roc_curve(y_test_bin_knn[:, i], y_prob_knn[:, i])
    roc_auc_knn[i] = auc(fpr_knn[i], tpr_knn[i])
    
    # Plotting ROC curve for class i logistic regression.
    ax = plt.subplot(1, 3, i + 1)
    plt.plot(fpr_log[i], tpr_log[i], color='darkorange', lw=lw, label='ROC curve Logistic (area = %0.2f)' % roc_auc_log[i])
    plt.plot(fpr_gnb[i], tpr_gnb[i], color='red', lw=lw, label='ROC curve GaussianNB (area = %0.2f)' % roc_auc_gnb[i])
    plt.plot(fpr_knn[i], tpr_knn[i], color='green', lw=lw, label='ROC curve KNN (area = %0.2f)' % roc_auc_knn[i])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC - Class '+str(i))
    plt.legend(loc="lower right")

plt.savefig('roc.png', dpi=600)
plt.show()
```
