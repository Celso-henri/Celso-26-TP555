```python
# Import all necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
```

```python
#vetor de exemplos
x = np.zeros((9,2))
x[0] = 1,4
x[1] = 4,3
x[2] = 4,5
x[3] = 3,6
x[4] = 6,7 
x[5] = 3,3
x[6] = 2,5
x[7] = 2,2
x[8] = 2,3

#vetor de centr√≥ides

c0 = np.zeros((1,2))
c0[0] = 5,3

c1 = np.zeros((1,2))
c1[0] = 1,3

c2 = np.zeros((1,2))
c2[0] = 3,4

C = np.zeros((3,2))
C[0] = 5,3
C[1] = 1,3
C[2] = 3.4
```

```python
#Figura mostrando os dados de treinamento
fig = plt.figure(figsize=(5,5))
plt.plot(x[:,0], x[:,1], '.', label='Exemplos')
plt.plot(c0[:,0], c1[:,1],'rx', label='Centr√≥ide 0')
plt.plot(c1[:,0], c1[:,1],'bx', label='Centr√≥ide 1')
plt.plot(c2[:,0], c1[:,1],'gx', label='Centr√≥ide 2')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.show()
```

```python
#Instanciando um objeto da classe KMeans
kmeans = KMeans(n_clusters=3, init=C)

#Treinando o modelo
clusters = kmeans.fit_predict(x)

#Imprimindo os Centr√≥ides √≥timos
for i in range(0,3):
    print('Centroid %d: (%1.2f, %1.2f)' % (i,kmeans.cluster_centers_[i][0],kmeans.cluster_centers_[i][1]))

#Imprimindo o n√∫mero de itera√ß√µes necess√°rias para se treinar o modelo
print('N√∫mero de itera√ß√µes=',kmeans.n_iter_)

#Imprimindo as classifica√ß√µes de cada exemplo
print(kmeans.labels_)
```

```python
#Figura com os dados de treinamento indicando √† que clusters cada um deles pertence al√©m de mostrar os centr√≥ides encontrados.

colors = cm.nipy_spectral(clusters.astype(float) / 3)
plt.scatter(x[:, 0], x[:, 1], marker='.', s=30, lw=1, alpha=0.7,
                c=colors, edgecolor='k')

# Labeling the clusters
centers = kmeans.cluster_centers_
# Draw white circles at cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='o',
            c="white", alpha=1, s=200, edgecolor='k')

for i, c in enumerate(centers):
    plt.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                s=50, edgecolor='k')
    
plt.xlabel("Espa√ßo de features para a primeira feature")
plt.ylabel("Espa√ßo de features para a segunda feature")

plt.show()
```
