```python
# Import all necessary libraries.
import numpy as np
import tensorflow as tf
import matplotlib
import matplotlib.pyplot as plt
```

```python
print(tf.__version__)
```

```python
# Create the dataset: XOR Logic Function
x = np.array([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
y = np.array([[0.], [1.], [1.], [0.]])
```

```python
# To make this notebook's output stable across runs
def reset_graph(seed=97):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

# Gradient descent Optimizer

```python
reset_graph()

n_epochs = 40000
learning_rate = 0.085

X = tf.placeholder(tf.float32, shape=(4, 2), name="X")
Y = tf.placeholder(tf.float32, shape=(4, 1), name="y")
W = tf.Variable(tf.random_uniform([2, 2], -1.0, 1.0), name="W")
w = tf.Variable(tf.random_uniform([2, 1], -1.0, 1.0), name="w")
B = tf.Variable([1., 1.], name="B")
b3 = tf.Variable([1.], name="b3")

y_pred = tf.sigmoid(tf.matmul(tf.sigmoid(tf.matmul(X,W)+B),w + b3))

with tf.name_scope("loss") as scope:
    error = y_pred - Y
    mse = tf.reduce_mean(tf.square(error), name="mse")

optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(mse)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(n_epochs):
        sess.run(training_op, feed_dict={X: x, Y: y})
        if epoch % 100 == 0:
            print("Epoch", epoch, "MSE =", mse.eval(feed_dict={X: x, Y: y}))
        
```

# Adam Optimizer

```python
reset_graph()

n_epochs = 40000
learning_rate = 0.085

X = tf.placeholder(tf.float32, shape=(4, 2), name="X")
Y = tf.placeholder(tf.float32, shape=(4, 1), name="y")
W = tf.Variable(tf.random_uniform([2, 2], -1.0, 1.0), name="W")
w = tf.Variable(tf.random_uniform([2, 1], -1.0, 1.0), name="w")
B = tf.Variable([1., 1.], name="B")
b3 = tf.Variable([1.], name="b3")

y_pred = tf.sigmoid(tf.matmul(tf.sigmoid(tf.matmul(X,W)+B),w + b3))

with tf.name_scope("loss") as scope:
    error = y_pred - Y
    mse = tf.reduce_mean(tf.square(error), name="mse")

optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(mse)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(n_epochs):
        sess.run(training_op, feed_dict={X: x, Y: y})
        if epoch % 100 == 0:
            print("Epoch", epoch, "MSE =", mse.eval(feed_dict={X: x, Y: y}))
```
